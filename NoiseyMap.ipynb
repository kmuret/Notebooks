```python
import folium
import pandas as pd
import folium.plugins

import branca
import branca.colormap as cm


for_map = pd.read_csv('noiselog220125_568.csv', sep=',')


#hm_wide = HeatMap( list(zip(for_map.lat.values, for_map.lon.values, for_map.Power.values)),
#                   min_opacity=0.2,
#                   radius=17, blur=15, 
#                   max_zoom=1, 
#                )


#data = [list(zip(for_map.lat.values, for_map.lon.values, for_map.Power.values))]
 
#df = pd.DataFrame(data, columns=['latitude','longitude','power'])

x_start = (for_map.lat.max() + for_map.lat.min()) / 2
y_start = (for_map.lon.max() + for_map.lon.min()) / 2
start_coord = (x_start, y_start)


colormap = cm.LinearColormap(colors=['white','red'], index=[for_map.Power.min(),for_map.Power.max()],vmin=for_map.Power.min(),vmax=for_map.Power.max())

map = folium.Map(location=start_coord, zoom_start=15)


lat = for_map.lat.values
lon = for_map.lon.values
pow = for_map.Power.values


for loc, p in zip(zip(lat, lon), pow):
    folium.Circle(
        location=loc,
        radius=10,
        fill=True,
        color=colormap(p),
        fill_opacity=0.7
    ).add_to(map)

map.add_child(colormap)

display(map)
```
